# 2. Basic Agents + Model Context Protocol (MCP)
Adds **MCP concepts** to the previous architecture.
MCP standardizes how models discover and use tools.

```python
# Simple MCP-style tool registry

class MCPServer:

    def __init__(self):
        self.tools = {}

    def register_tool(self, name, func):
        self.tools[name] = func

    def call_tool(self, name, *args):
        return self.tools[name](*args)


# Example tools

def weather(city):
    data = {"mumbai": "Hot", "london": "Rainy"}
    return data.get(city.lower(), "Unknown")

def wiki(topic):
    return f"Wikipedia summary for {topic}"


server = MCPServer()

server.register_tool("weather", weather)
server.register_tool("wiki", wiki)

server.tools
```

```python
# MCP Client

class MCPClient:

    def __init__(self, server):
        self.server = server

    def use_tool(self, tool_name, arg):
        return self.server.call_tool(tool_name, arg)


client = MCPClient(server)

client.use_tool("weather", "Mumbai")
```

```python
# Agent using MCP tools

def agent(query):

    if "weather" in query:
        city = query.split()[-1]
        return client.use_tool("weather", city)

    if "wiki" in query:
        topic = query.split()[-1]
        return client.use_tool("wiki", topic)

    return "No tool matched"

agent("weather Mumbai")
```

```python
# Memory + Middleware integrated

memory = []

def middleware(agent_func):

    def wrapper(query):
        print("Query:", query)
        result = agent_func(query)
        memory.append((query, result))
        return result

    return wrapper

agent_with_middleware = middleware(agent)

agent_with_middleware("wiki python")

memory
```
