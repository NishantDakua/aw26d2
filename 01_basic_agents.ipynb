# 1. Basic Agent Concepts (No Framework)
Topics:
- Tool Calling
- Agent Memory
- Dynamic Agent Routing
- Agent Middleware
- Third Party Tools

This notebook shows **how agents work internally without frameworks**.

```python
# Example third‑party tool (simulated)
# In real applications this would call an external API

def search_tool(query):
    """Simulated search API"""
    data = {
        "python": "Python is a programming language.",
        "ai": "AI stands for Artificial Intelligence."
    }
    return data.get(query.lower(), "No results found")
```

```python
# Tool calling example

def tool_calling_agent(query):
    if "search" in query:
        topic = query.replace("search", "").strip()
        return search_tool(topic)
    return "Agent could not determine tool"

print(tool_calling_agent("search python"))
```

```python
# Simple Agent Memory

class AgentMemory:
    def __init__(self):
        self.history = []

    def add(self, message):
        self.history.append(message)

    def get(self):
        return self.history

memory = AgentMemory()
memory.add("User asked about Python")
memory.add("Agent responded with definition")

memory.get()
```

```python
# Dynamic Agent Routing

def math_agent(query):
    return f"Math agent solving: {query}"

def research_agent(query):
    return search_tool(query)

def router(query):
    if "calculate" in query:
        return math_agent(query)
    return research_agent(query)

router("python")
```

```python
# Agent Middleware

def logging_middleware(func):
    def wrapper(query):
        print("LOG: received query ->", query)
        result = func(query)
        print("LOG: response ->", result)
        return result
    return wrapper

@logging_middleware
def simple_agent(query):
    return router(query)

simple_agent("python")
```
